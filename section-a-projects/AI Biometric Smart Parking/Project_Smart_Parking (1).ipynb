```python
# STEP 2: Import Required Libraries
import os
import gdown
import zipfile
import pandas as pd
from matplotlib import pyplot as plt
from google.colab import files
from IPython.display import display
```

```python
# STEP 3: Download and Extract the Dataset
file_id = "15FthdQ1TVjB29_4GWodngor2fDKwgsJL"
zip_path = "plates_dataset.zip"
dataset_dir = "registered_plates"

gdown.download(f"https://drive.google.com/uc?id={file_id}", zip_path, quiet=False)

os.makedirs(dataset_dir, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(dataset_dir)

print("âœ… Dataset successfully downloaded and unzipped.")
```

```python
import easyocr

reader = easyocr.Reader(['en'])

def extract_plate_number(img_path):
    print(f"ğŸ“· Reading with EasyOCR: {img_path}")

    results = reader.readtext(img_path, detail=0, paragraph=False)

    # Combine and clean result
    if results:
        plate_text = ' '.join(results).upper()
        print(f"ğŸ“ EasyOCR Output: {plate_text}")
        return plate_text
    else:
        print("âš ï¸ EasyOCR failed to detect any text.")
        return "NOT_DETECTED"
```

```python
# STEP 5: Process Dataset Images and Extract Plates
reader = easyocr.Reader(['en'])

print("\nğŸ” Scanning dataset images...")

registered_plates_data = []

# âœ… Change: Recursively search for image files in all subdirectories
image_extensions = ('.jpg', '.jpeg', '.png')
files_in_dataset = []

for root, dirs, files in os.walk(dataset_dir):
    for file in files:
        if file.lower().endswith(image_extensions):
            full_path = os.path.join(root, file)
            files_in_dataset.append(full_path)

print(f"ğŸ“‚ Found {len(files_in_dataset)} image files in dataset.")

if not files_in_dataset:
    print("âŒ No image files found. Please check dataset structure.")

# âœ… Loop over full image paths found
for image_path in files_in_dataset:
    print(f"\nğŸ“¸ Processing image: {image_path}")

    plate_text = extract_plate_number(image_path)
    print(f"ğŸ“ OCR Output: {plate_text}")

    registered_plates_data.append({
        'Filename': os.path.basename(image_path),
        'Plate': plate_text
    })

    print(f"âœ… {os.path.basename(image_path)} âœ Plate Detected: '{plate_text}'")
```

```python
# STEP 6: Create DataFrame
df_registered = pd.DataFrame(registered_plates_data)

print("\nğŸ§¾ Columns in DataFrame:", df_registered.columns.tolist())

if df_registered.empty:
    print("âš ï¸ No plates found! Check OCR or dataset.")
else:
    print(f"âœ… Total registered plates: {len(df_registered)}")
    display(df_registered)
```

```python
# STEP 7: Convert to List for Matching
registered_plates = df_registered['Plate'].tolist()
```
